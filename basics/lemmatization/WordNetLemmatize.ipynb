```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
''' pos values:
noun-n
verb-v
adverb-r
adjective-a'''
```

```python
lemmatizer.lemmatize('going')
```

```python
lemmatizer.lemmatize('going',pos='n')
```

```python
lemmatizer.lemmatize('going',pos='v')#we can see as going is changed to go only when we mention v[verb] in pos
```

```python
words=['going','sportingly','fairly','eating']
for word in words:
    print(lemmatizer.lemmatize(word,pos='v'))
```
